```python
import requests
import pandas as pd
import os, subprocess
```

```python
params = "cnt:switzerland" 
url="https://www.xeno-canto.org/api/2/recordings?query="+params

r=requests.get(url, headers={"Content-Type":"json"})
resp = r.json()
print(resp)
```

```python
df = pd.DataFrame(resp["recordings"])
df.keys()
```

```python
urls = []
for index in df.index:
  base=df.sono[index]['small'][:43]
  #print(df.sono[0]['small'])
  filename=df['file-name'][index]
  #print(df['file-name'][0])
  url='http:'+base+'/'+filename
  urls.append(url)
df['urls']=urls
```

```python
# download all mp3 files from df['urls']
url = df['urls'][6].replace(" ", "%20")
files_list = []
# avoid to load empty urls 
if url.startswith('http'):
  resp1 = requests.get(url, allow_redirects=True)
  data = resp1.content
  pathtofile, file_extension = os.path.splitext(url)
  file_name = pathtofile.replace("%20", "_")[49:] + file_extension 
  folder = pathtofile[38:48]
  # check if file already there
  if os.path.isfile(file_name):
    print('file already exist')
  # write file inside folder
  else:
    a_file = os.path.join(folder, file_name)
    files_list.append(a_file)
    # print(item);
    with open(a_file, 'wb') as f:
      f.write(data)
```
